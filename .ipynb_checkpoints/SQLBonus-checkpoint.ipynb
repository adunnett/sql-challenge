```python
import pandas as pd
from sqlalchemy import create_engine
from config import Password
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats

engine = create_engine(f"postgresql://postgres:{Password}@localhost:5432/EmployeeSQL")
connection = engine.connect()
```

```python
employees = pd.read_sql("SELECT * FROM Employees", connection)
salaries = pd.read_sql("SELECT * FROM Salaries", connection)
titles = pd.read_sql("SELECT * FROM Titles", connection)
```

```python
# Create a histogram to visualize the most common salary ranges for employees.
plt.hist(salaries['salary'], bins=20, color='#86bf91', label='Salary')
plt.axvline(salaries['salary'].mean(), color='k', linestyle='dashed', linewidth=1, label="Salary Mean")
plt.title("Common Salary Ranges")
plt.xlabel("Salary", labelpad=20, weight='bold', size=12)
plt.ylabel("Number of Employees", labelpad=20, weight='bold', size=12) 
plt.legend()
plt.savefig("Common_Salary_Ranges_Hist.png")
```

```python
# Create a bar chart of average salary by title.
merge1 = pd.merge(salaries, employees, on='emp_no', how='inner')
titles.rename(columns={'title_id' : 'emp_title'}, inplace=True)
merge2 = pd.merge(employees, titles, on='emp_title', how='inner')
merge3 = pd.merge(merge1, merge2, on='emp_no', how='inner')
merge4 = merge3.groupby(['title']).mean()['salary']
```

```python
ax = merge4.plot.bar(legend=False, color='#86bf91')
plt.title('Average Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.savefig("avg_salary_by_title.png")
```
